```python
import re
def clean_page(page):
    page = re.sub(r'<!--(.*?)-->', '', page)

    # Remove &nbsp;
    page = page.replace('&nbsp;', '')
    soup = BeautifulSoup(page, 'html.parser')

    # Remove CSS styles
    for style in soup(["style", "link", "script"]):
        style.extract()

    # Find all text elements
    text_elements = soup.find_all(text=True)

    # Extract the text
    text_data = ' '.join(text_elements)
    return text_elements
```
