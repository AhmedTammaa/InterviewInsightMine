```python
def clean(html_content):

    # Assuming 'html_content' contains the HTML data from your JSON file

    soup = BeautifulSoup(html_content, 'html.parser')

    # Initialize lists to store the extracted data
    company_names = []
    roles = []
    dates = []
    questions = []

    # Find all the 'mb borderBot' divs
    divs = soup.find_all('div', class_='mb borderBot')

    # Iterate over each div and extract the desired data
    for div in divs:
        # Extract the company name from the 'alt' attribute of the 'img' tag
        img = div.find('img')
        if img is not None:
            company_name = img.get('alt')
            company_names.append(company_name)
        else:
            company_names.append('unkown')
        # Extract the role from the text within the 'a' tag inside the 'span' tag
        spans = div.find_all('span', class_='m-0 css-1jlgt0v edupdmz5')
        if (len(spans)>=2):
            role = spans[0].find('a').text
            date = spans[1].text
            #print(company_name)
            roles.append(role)
            dates.append(date)
        #role = div.find('span', class_='m-0 css-1jlgt0v edupdmz5').find('a').text
        #roles.append(role)

        # Extract the date from the text within the 'span' tag
        #date = div.find('span', class_='m-0 css-1jlgt0v edupdmz5').text
        #dates.append(date)

        # Extract the question from the text within the 'h3' tag
        question = div.find('h3', class_='css-11euy82 edupdmz3').text
        questions.append(question)

    # Create a dataframe from the extracted data
    data = {
        'Company Name': company_names,
        'Role': roles,
        'Date': dates,
        'Question': questions
    }
    #print(roles)
    
    df = pd.DataFrame(data)
    #df['Role'] = df['Role'].str.replace(' was asked...', '',regex=True)
    # Display the dataframe
    return df
```
